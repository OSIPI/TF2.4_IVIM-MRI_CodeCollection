# Visualizing test results

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
```

```python
test_root = '../tests/results'
contributions = os.listdir(test_root)
units = {'f': '%', 'D': r'mm$^2$/s', 'Dp': r'mm$^2$/s'}
scale = {'f': 1e2, 'D': 1e3, 'Dp': 1e3}
```

```python
fig, axes = plt.subplots(3, 1, figsize=(1*5,3*5))
for contribution in contributions:
    df = pd.read_csv(os.path.join(test_root, contribution))
    for par, ax in zip (['f', 'D', 'Dp'], axes):
        ax.plot((df[f'{par}_fit']-df[f'{par}_true'])*scale[par],'o')
        ax.set_ylabel(f'{par}_fit - {par}_true [{units[par]}]')
```

```python
fig, axes = plt.subplots(3, 1, figsize=(1*5,3*5))
for contribution in contributions:
    df = pd.read_csv(os.path.join(test_root, contribution))
    for par, ax in zip (['f', 'D', 'Dp'], axes):
        ax.plot(df[f'{par}_true']*scale[par], df[f'{par}_fit']*scale[par],'o')
        ax.set_ylabel(f'{par}_fit [{units[par]}]')
        ax.set_xlabel(f'{par}_true [{units[par]}]')
```
